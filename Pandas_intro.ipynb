```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

def new_column(x):
    if x <=10000:
        return 'A'
    elif x <=100000 and x >10000:
        return 'B'
    elif x <=200000 and x >100000:
        return 'C'
    elif x <=400000 and x >200000:
        return 'D'
    elif x <=700000 and x >400000:
        return 'E'
    elif x > 700000:
        return 'F'
    
df['new_col'] = df['LIMIT_BAL'].map(new_column)

df[['LIMIT_BAL', 'new_col']].head(17).T
```

```python
# (8) построить среднее значение кредитного лимита для каждого вида образования 
# и для каждого пола
# график необходимо сделать очень широким (на весь экран)
#TODO 
df_1 = df[['SEX', 'EDUCATION', 'LIMIT_BAL']]
df_1['SEX'].replace(to_replace={1:'male', 2:'female'}, inplace=True) 
df_1['EDUCATION'].replace(to_replace={1:'graduate school', 2:'university', 3:'high school', 
                                      4:'others', 0:'others', 5:'others', 6:'others'}, inplace=True) 

df_pvt_tbl = df_1.pivot_table(values='LIMIT_BAL', index='EDUCATION', columns='SEX', aggfunc='mean', )

_, ax = plt.subplots(figsize=(20,8))

df_pvt_tbl.plot(kind='bar', ax=ax)
plt.show()
```

```python
# (9) построить зависимость кредитного лимита и образования только для одного из полов

#TODO

df_2 = df_1[df_1['SEX'] == 'male']
df_2_pvt_tbl = df_2.pivot_table(values='LIMIT_BAL', index='EDUCATION', aggfunc='mean')

_, ax = plt.subplots(figsize=(15,8))

df_2_pvt_tbl.plot(kind='bar', ax=ax)
plt.show()
```
